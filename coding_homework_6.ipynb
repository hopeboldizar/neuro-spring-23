```python
#2 Make a Code Cell that imports numpy as np and matplotlib.pyplot as plt
# Import our plotting package from matplotlib
import matplotlib.pyplot as plt

# Import numpy below
import numpy as np

# Add your print() statement below
print('Packages imported!')
```

```python
#3 create numpy arrays for the imported data
url1 = 'https://raw.githubusercontent.com/hopeboldizar/neuro-spring-23/main/acc_hits_hw6%20(1).csv'
acc_hits = np.loadtxt(url1, delimiter = ',')
url2 = 'https://raw.githubusercontent.com/hopeboldizar/neuro-spring-23/main/acc_fa_hw6%20(1).csv'
acc_fa = np.loadtxt(url2, delimiter = ',')
url3 = 'https://raw.githubusercontent.com/hopeboldizar/neuro-spring-23/main/lc_hits_hw6%20(1).csv'
lc_hits = np.loadtxt(url3, delimiter = ',')
url4 = 'https://raw.githubusercontent.com/hopeboldizar/neuro-spring-23/main/lc_fa_hw6%20(1).csv'
lc_fa = np.loadtxt(url4, delimiter = ',')
```

```python
#4 print the shape of each array
print(acc_hits.shape)
print(acc_fa.shape)
print(lc_hits.shape)
print(lc_fa.shape)
```

```python
#6 create time variable
time =np.linspace(0,4, num=8001)
print(time)
```

```python
#7 create expression to determine sampling rate
8001/4
```

```python
#8 plotting first trial of acc_hits and acc_fa
fig, ax = plt.subplots(1,2,figsize=(15,5))

ax[0].plot(time,acc_hits[0], color= "red")
ax[0].plot(time,lc_hits[0], color= "blue")
ax[1].plot(time,acc_fa[0], color= "red")
ax[1].plot(time,lc_fa[0], color= "blue")

ax[0].set_ylabel('LFP')
ax[0].set_xlabel('time')
ax[1].set_ylabel('LFP')
ax[1].set_xlabel('time')
```

```python
#10 computing coherence
# Compute the Fourier transforms
xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_hits])  
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_hits])  

# Compute the spectra
T = time[-1]
dt = time[1] - time[0] 
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0) 
Sxy = scale * (xf * yf.conj()).mean(0)  

# Compute the coherence.
cohr_hits = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))              
N = lc_hits.shape[1]
f = np.fft.rfftfreq(N, dt)                    
plt.plot(f, cohr_hits.real, color = 'red')                      
plt.xlim([0, 50])                           
plt.ylim([0, 1])                           
plt.xlabel('Frequency [Hz]')                
plt.ylabel('Coherence')
plt.title('Coherence between two electrodes')
```

#11 which freqeuency band has the most phase coherence?

```python
#12 compute coherence for acc_fa and lc_fa
# Compute the Fourier transforms
xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_fa])  
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_fa])  

# Compute the spectra
T = time[-1]
dt = time[1] - time[0] 
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0) 
Sxy = scale * (xf * yf.conj()).mean(0)  

# Compute the coherence.
cohr_fa = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))              
N = lc_fa.shape[1]
f = np.fft.rfftfreq(N, dt)                    
plt.plot(f, cohr_fa.real, color = 'blue')                      
plt.xlim([0, 50])                           
plt.ylim([0, 1])                           
plt.xlabel('Frequency [Hz]')                
plt.ylabel('Coherence')
plt.title('Coherence between two electrodes')
```

```python
# plot coherence for fa and acc on one plot
plt.plot(f, cohr_hits.real, color = 'red')
plt.plot(f, cohr_fa.real, color = 'blue')
plt.xlim([0,50])
plt.ylim([0,1])
plt.xlabel('Frequency [Hz]')                
plt.ylabel('Coherence')
plt.title('Coherence between two electrodes')
```
