```python
#2 Make a Code Cell that imports numpy as np and matplotlib.pyplot as plt
# Import our plotting package from matplotlib
import matplotlib.pyplot as plt

# Import numpy below
import numpy as np

# Add your print() statement below
print('Packages imported!')
```

```python
#3 import files as numpy arrays
url1 = 'https://raw.githubusercontent.com/hopeboldizar/neuro-spring-23/main/acc_hits_hw5.csv'
acc_hits = np.loadtxt(url1, delimiter = ',')
url2 = 'https://raw.githubusercontent.com/hopeboldizar/neuro-spring-23/main/acc_fa_hw5.csv'
acc_fa = np.loadtxt(url2, delimiter = ',')
```

```python
#4 print shape of array
print(acc_hits.shape)
print(acc_fa.shape)
```

```python
#6 creating time variable
time =np.linspace(0,4, num=8001)
print(time)
```

```python
#7 sampling rate
8001/4
```

```python
#8 plotting first trial of acc_hits and acc_fa
fig, ax = plt.subplots(2,2,figsize=(15,5))

ax[0,0].plot(time,acc_hits[0], color= "skyblue")
ax[0,1].plot(time,acc_fa[0], color= "hotpink")

plt.ylabel('LFP')
plt.xlabel('time')
```

##9 describing dominant oscillations
In the acc_hits plot, there are 5 peaks over the course of 4 seconds, so the dominant oscillation is 1.25. In the acc_fa plot, there are 4 clear peaks over 4 seconds, so the dominant oscillation is 1. The values are close, but they do differ slightly. 

```python
#10 calculating ERP
plt.plot(time, acc_hits.mean(0))        # Plot the ERP of condition A
plt.xlabel('Time [s]')           # Label the axes
plt.ylabel('LFP Voltage [Î¼V]')
plt.title('ERP of acc_hits')  # ... provide a title
plt.plot([0.25, 0.25], [-0.3,0.3], 'k', lw=2)
plt.hlines(0, time[0], time[-1], color='k')
plt.axvline(x=2, color='k')
```

#11 interpret ERP
There does not seem to be consistent activity in the LFP surrounding screen touch, as there does not appear to be much data or peaks in data near the line.

```python
#12 calculating ERP for acc_fa
plt.plot(time, acc_fa.mean(0))        
plt.xlabel('Time [s]')           
plt.ylabel('LFP Voltage [Î¼V]')

plt.plot(time, acc_hits.mean(0))        # Plot the ERP of condition A
plt.xlabel('Time [s]')           # Label the axes
plt.ylabel('LFP Voltage [Î¼V]')
plt.title('ERP of acc_hits(orange) and acc_fa(blue)')  


plt.plot([0.25, 0.25], [-0.3,0.3], 'k', lw=2)
plt.hlines(0, time[0], time[-1], color='k')
plt.axvline(x=2, color='k')
```

There does seem to be consistent activity in the LFP surrounding screen touch, as there is a trough in the data directly to the left of the vertical line, and a peak directly to the right.
