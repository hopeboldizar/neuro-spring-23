```python
#2 Make a Code Cell that imports numpy as np and matplotlib.pyplot as plt
# Import our plotting package from matplotlib
import matplotlib.pyplot as plt

# Import numpy below
import numpy as np

# Add your print() statement below
print('Packages imported!')
```

```python
#10 calculate ISI for the 4 arrays
ISIleft1 = np.diff(spikes_left_trial1)
ISIleft2 = np.diff(spikes_left_trial2)
ISIright1 = np.diff(spikes_right_trial1)
ISIright2 = np.diff(spikes_right_trial2)
```

```python
#11 make a histogram for ISI data
bins = np.arange(0, 2, 0.01) 
plt.hist(ISIleft1, bins)
plt.xlabel('ISI [s]')          
plt.ylabel('Counts')             
plt.title('Left Trial 1') 
```

```python
#11 make a histogram for ISI data
bins = np.arange(0, 2, 0.01) 
plt.hist(ISIleft2, bins)
plt.xlabel('ISI [s]')          
plt.ylabel('Counts')             
plt.title('Left Trial 2') 
```

```python
#11 make a histogram for ISI data
bins = np.arange(0, 2, 0.01) 
plt.hist(ISIright1, bins)
plt.xlabel('ISI [s]')          
plt.ylabel('Counts')             
plt.title('Right Trial 1') 
```

```python
#11 make a histogram for ISI data
bins = np.arange(0, 2, 0.01) 
plt.hist(ISIright2, bins)
plt.xlabel('ISI [s]')          
plt.ylabel('Counts')             
plt.title('Right Trial 2') 
```

#12 interpret histograms of ISI data


*   All the graphs show higher and more concentrated ISI data towards the left of the graph, indicating the neurons tend to fire in bursts. 
*   For both left and right, the first several bins seem to be where the more "bursting" occurs



```python
#13 calculating firing rate across time

binsize = 1                                                                  
fr_bins1 = np.arange(0,30,binsize)                                            
fr_left1 = np.zeros(len(fr_bins1)-1)                                                                        
for i in range(len(fr_bins1)-1):                                               
  spk_index = (spikes_left_trial1 > fr_bins1[i]) & (spikes_left_trial1 < fr_bins1[i+1])           
  true_values = spikes_left_trial1[spk_index]                                          
  fr_left1[i] = (len(true_values))/binsize                                     

binsize = 1                                                               
fr_bins1 = np.arange(0,30,binsize)                                             
fr_right1 = np.zeros(len(fr_bins1)-1)                                                                             
for i in range(len(fr_bins1)-1):                                             
  spk_index = (spikes_right_trial1 > fr_bins1[i]) & (spikes_right_trial1 < fr_bins1[i+1])          
  true_values = spikes_right_trial1[spk_index]                                          
  fr_right1[i] = (len(true_values))/binsize 

binsize = 1                                                                  
fr_bins2 = np.arange(0,30,binsize)                                            
fr_left2 = np.zeros(len(fr_bins2)-1)                                                                        
for i in range(len(fr_bins2)-1):                                               
  spk_index = (spikes_left_trial2 > fr_bins2[i]) & (spikes_left_trial2 < fr_bins2[i+1])           
  true_values = spikes_left_trial2[spk_index]                                          
  fr_left2[i] = (len(true_values))/binsize                                     

binsize = 1                                                               
fr_bins2 = np.arange(0,30,binsize)                                             
fr_right2 = np.zeros(len(fr_bins2)-1)                                                                             
for i in range(len(fr_bins2)-1):                                             
  spk_index = (spikes_right_trial2 > fr_bins2[i]) & (spikes_right_trial2 < fr_bins2[i+1])          
  true_values = spikes_right_trial2[spk_index]                                          
  fr_right2[i] = (len(true_values))/binsize 

fig, ax = plt.subplots(1,2,figsize=(15,5))

ax[0].plot(fr_bins1[:-1], fr_left1, color = 'hotpink')
ax[0].plot(fr_bins1[:-1], fr_right1, color = 'lightblue')
ax[0].set_xlabel('Time [s]')
ax[0].set_ylabel('Firing Rate [Hz]')
ax[0].set_title('trial 1')
ax[1].plot(fr_bins2[:-1], fr_left2, color = 'hotpink')
ax[1].plot(fr_bins2[:-1], fr_right2, color = 'lightblue')
ax[1].set_xlabel('Time [s]')
ax[1].set_ylabel('Firing Rate [Hz]')
ax[1].set_title('trial 2')
```
